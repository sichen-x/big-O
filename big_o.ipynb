Write a script that produces a single plot file named big_o.png with the following functions

    O(1)
    O(log n)
    O(n log n )
    O(n)
    O(n^2)
    O(2^n)

The last two functions should be plotted in logarithmic scale.

```python
import math
import numpy as np
from bokeh.io import output_notebook
output_notebook()
from bokeh.plotting import figure, output_file, show
```

```python
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.plotting import figure, output_file, show
```

```python
X = [ i for i in range(1, 50) ]

TOOLS = "pan,wheel_zoom,box_zoom,reset,save"
p = figure(height=600, width=800, tools=TOOLS)

p.title.text = 'Big O functions'
p.title.align = 'center'
p.xaxis.axis_label = 'n'
p.yaxis.axis_label = 'f(n)'
p.line(x=X, y=[ 1 for i in X], line_width=2, legend='O(1)', color='red')
p.line(x=X, y=[ math.log2(i) for i in X], line_width=2, legend='O(log n)', color='blue')
p.line(x=X, y=[ i * math.log2(i) for i in X], line_width=2, legend='O(n log n)', color='orange')
p.line(x=X, y=[ i for i in X], line_width=2, legend='O(n)', color='green')
p.line(x=X, y=[ math.log(i * i) for i in X], line_width=2, legend='O(n^2)', color='gray')
p.line(x=X, y=[ math.log(math.pow(2, i)) for i in X], line_width=2, legend='O(2^n)', color='yellow')
p.legend.location = "top_left"
show(p)
```
